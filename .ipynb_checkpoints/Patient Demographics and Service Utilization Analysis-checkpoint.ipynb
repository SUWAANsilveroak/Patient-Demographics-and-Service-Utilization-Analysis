```python
# Checking how many Billing Amount is there which is equal to zero
equal_to_zero_amounts = fdf[fdf['Billing Amount']==0]['Billing Amount']
print(len(equal_to_zero_amounts))
```

```python
# Checking how many Billing Amount is there which is less than zero
negative_amounts = fdf[fdf['Billing Amount']<0]['Billing Amount']
print(len(negative_amounts))
```

```python
# Luckily there are no values which are equal to 0 but there are some negative amount due to type error 
# so we need to replace those negative amount into positive 
```

```python
# Function to convert amount into rounded and positive
def negative_to_positive(amount):
    
    # Rounding off the billing amount before coverting them
    rounded_amount= round(amount)
        
    # now coverting negative amounts in positive
    if rounded_amount<0 :
        rounded_amount= abs(rounded_amount)
            
    return rounded_amount

# applying the function in our data frame 
fdf['Billing Amount']=fdf['Billing Amount'].apply(negative_to_positive)

# Now Billing amount is cleaned
fdf.head()
```

#### There are 2 columns "Date of Admission" and "Discharge Date" we need to convert them into datetime 

```python
# Am using built-in function pd.to_datetime to covert them
fdf[["Date of Admission", "Discharge Date"]] = fdf[["Date of Admission", "Discharge Date"]].apply(pd.to_datetime)
# Lets check
fdf.info()
```

#### Date of Admission and Discharge Date was Object now it's datetime64

```python
# Lets create a new column total days for future analysis 
fdf['Total_days']=(fdf['Discharge Date']-fdf['Date of Admission']).dt.days
fdf.head()
```

```python
fdf.describe()
```

### The min Billing amount is 9 dollars which can't be correct so we gonna take median of 25% percentile of the data and remove the rows which is below median of 25% percentile 

(This decision depends on managment or people with domain knowledge i used this method to work with proper min vlaue to work with)

```python
# The percentile as of now which is 13244.
billing_count= []
for num in fdf['Billing Amount']:
    if num <=13244:
       billing_count.append(num) 


# Finding the median value
import statistics
billing_median = statistics.median(billing_count)

# Updating fdf by filtering out rows where Billing Amount is less than 7074.5
fdf = fdf[fdf['Billing Amount'] >= 7074.5]
fdf.shape
```

```python
# lets count
def count_freq(Gender):
    male = []
    female = []
    
    # Loop to classify gender into male and female lists
    for cat in Gender:
        if cat == 'Male':
            male.append(cat)
        else:
            female.append(cat)
    
    # Print counts
    print("Male count:", len(male))
    print("Female count:", len(female))


count_freq(fdf['Gender'])
# There are minor diffrence 
```

#### Observation :- Male and female patient are almost equal

```python
# lets find out wich medical condition is in high in frequesncy

def count_frequency(df, column_name):
    """
    This function counts the frequency of unique values in a specified column of a DataFrame.
    
    Parameters:
    df (pandas.DataFrame): The DataFrame containing the data.
    column_name (str): The column name for which to count the frequency of values.

    Returns:
    pandas.Series: Frequency counts of unique values in the specified column.
    """
    if column_name in df.columns:
        condition_counts = df[column_name].value_counts()
        return condition_counts
    else:
        print(f"Column '{column_name}' does not exist in the DataFrame.")
        return None
```

```python
# Checking the function
Medical_condition_count=count_frequency(fdf,'Medical Condition')
print(Medical_condition_count)
```

```python
# Let,s plot the medical condition 
plt.figure(figsize=(10,6))
sns.barplot(x=Medical_condition_count.index, y=Medical_condition_count.values, palette='Set2')

plt.title("Most frequent Medical Condition")
plt.xlabel("Medical Conditon")
plt.ylabel("frequency")
plt.show()
```

```python
# lets plot pie chart for better insight

plt.figure(figsize=(10,6))
plt.pie(Medical_condition_count,labels=Medical_condition_count.index,autopct='%1.1f%%',startangle=90, colors=sns.color_palette("Set2"))

plt.title("Medical Condition Insight")
plt.axis('equal')
plt.show()
```

```python
# Now lets find out which medical condition cost the highest 

# To find out first lets create the new column "Charge Per Day"

fdf['Charge Per Day'] = fdf['Billing Amount'] / fdf['Total_days']
fdf.head()
```

```python
# Lets check the diostribution of Charge per day
plt.figure(figsize=(10, 6))
sns.histplot(fdf['Charge Per Day'], bins=30, kde=True, color='blue')
plt.title('Distribution of Charge Per Day')
plt.xlabel('Charge Per Day')
plt.ylabel('Frequency')
plt.show()
```

### Observbation:- Charge per day have Right-Skewed Distribution for normal distribution we gonna use Log Transformation (Not stricly required but for the future use)

```python
fdf['Log Charge Per Day'] = np.log(fdf['Charge Per Day'] + 1)
sns.histplot(fdf['Log Charge Per Day'], kde=True, color='blue')
```

```python
fdf.head()
```

```python
# now lets Calculate the Median Charge per Day for Each Medical Condition 
# Because there are possibility  that different hospital charge different for the same  medical condition

median_charge_per_condition = fdf.groupby('Medical Condition')['Charge Per Day'].median().reset_index()
```

```python
# Lets plot the barplot for better analysis

plt.figure(figsize=(10, 6))
bars = sns.barplot(data=median_charge_per_condition, x='Medical Condition', y='Charge Per Day', palette='RdYlGn', width=0.5)

# Add data labels
for bar in bars.patches:
    bars.annotate(f'{bar.get_height():.2f}', (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                  ha='center', va='bottom', color='black', fontsize=10)

plt.title('Median Charge Per Day by Medical Condition') 
plt.xlabel('Medical Condition')
plt.ylabel('Charge Per Day') 
plt.xticks(rotation=45)

plt.ylim(0, 2300)
plt.show()
```

### Diabetes cost the most compare to others  

```python
# Which medical condition tooks higher day to heal 

# Let's Calculate the average total days for each medical condition
average_days_per_condition = fdf.groupby('Medical Condition')['Total_days'].mean().reset_index()

# Displaying the result
print(average_days_per_condition)
```

#### Observation:- Median value is  also same which indicates that there  is no outlier we gonna stick with mean for better result 

```python
# Lets plot the average days per medical condition
plt.figure(figsize=(10, 6))
bars = sns.barplot(x='Medical Condition', y='Total_days', data=average_days_per_condition, palette='coolwarm')

# Add annotations for each bar
for bar in bars.patches:
    bars.annotate(f'{bar.get_height():.2f}', 
                  (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                  ha='center', va='bottom', color='black', fontsize=10)

# Set the title and labels
plt.title('Average Days to Discharge by Medical Condition')
plt.xlabel('Medical Condition')
plt.ylabel('Average Total Days')

plt.xticks(rotation=45)
plt.ylim(0,18)
plt.show()
```

### Observation :- Compare to others Asthama took more time to discharge the patients

```python
# Lets find out the Top 5 Hospitals with highest incom

# Grouping the data by Hospital and calculating the total income for each
total_income_per_hospital = fdf.groupby('Hospital')['Billing Amount'].sum().reset_index()

# Sorting the hospitals by total income in descending order and selecting the top 5
top_5_hospitals = total_income_per_hospital.sort_values(by='Billing Amount', ascending=False).head(5)
print(top_5_hospitals)
```

```python
# Lets plot the figure 
plt.figure(figsize=(10, 6))
bars = sns.barplot(data=top_5_hospitals, x='Hospital', y='Billing Amount', palette='Blues_d')

# Annotation
for bar in bars.patches:
    bars.annotate(f'{bar.get_height():,.2f}', (bar.get_x() + bar.get_width() / 2, bar.get_height()), 
                  ha='center', va='bottom', color='black', fontsize=10)

plt.title('Top 5 Hospitals by Total Income', fontsize=16)
plt.xlabel('Hospital', fontsize=12)
plt.ylabel('Total Income ($)', fontsize=12)
plt.xticks(rotation=45)  

plt.show()
```
